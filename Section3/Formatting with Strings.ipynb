## Formatting with .format() method

```python
print('This is a string {}'.format('INSERTED'))
```

```python
print('The {} {} {}'.format('fox', 'brown', 'quick'))
```

```python
print('The {2} {1} {0}'.format('fox', 'brown', 'quick')) #supply numbers in order
```

```python
print('The {0} {0} {0}'.format('fox', 'brown', 'quick')) #supply numbers in order
```

```python
print('The {q} {b} {f}'.format(f='fox', b='brown', q='quick')) #using key words as variable hames and inserting
```

## Float formatting follows "{value:width.precision f}"

```python
result = 100/777
```

```python
result = 104.12345
```

```python
print("The result was {r}".format(r=result))
```

```python
print("The result was {r:7.3f}".format(r=result)) #Addiding width
```

## F-strings method 

```python
name = "Pranav"
```

```python
print('Hello, his name is {}'.format(name))
```

```python
print(f'Hello, his name is {name}') #with f-string method
```

Works with multiple variables

```python
name = "Pranav"
age = 15
```

```python
print(f'{name} is {age} years old')
```
